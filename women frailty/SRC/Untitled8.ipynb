**BMI Calculation**

```python
# Calculate BMI from weight & height

data['bmi'] = (703 * (data['weight']))/(data['height'] ** 2)
data['bmi'] = data['bmi'].round(2)

# Calculate grip strength in pounds
data['grip_strength_lbs'] = data['grip strength'] * 2.20462  # Conversion factor from KGS to LBS
data['grip_strength_lbs'] = data['grip_strength_lbs'].round(2)
# Calculate frailty index
# Got this from open source
data['frailty_index'] = (data['height'] / data['weight']) * data['age'] + (data['grip_strength_lbs'] / data['weight']) * data['age']
data['frailty_index'] = data['frailty_index'].round(2)

# Display the DataFrame with frailty index
print(data)
```

```python
data.to_csv("/content/cleaned_women_frailty1.csv")
```

**DATA VISUALIZATION**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Create a boxen plot to show the extended distribution
sns.boxenplot(x='frailty', y='frailty_index', data=data, palette='viridis')

# Overlay with a strip plot to show individual data points
sns.stripplot(x='frailty', y='frailty_index', data=data, color='k', alpha=0.5, jitter=True)  # 'k' for black, with some transparency and jitter for separation

plt.title('Frailty Index across Frailty Categories with Enhanced Distribution View')
plt.xlabel('Frailty Status')
plt.ylabel('Frailty Index')

# Save the figure in the adjusted file path
plt.savefig('/content/Snaps_of_plots/Frailty_Index_vs_Frailty_Category_Boxen_Strip.jpg')
plt.show()
```

**BMI Vs Frailty status**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
# Create a box plot to show the central tendency and dispersion
sns.boxplot(x='frailty', y='bmi', data=data, palette='Pastel1', width=0.5)

# Overlay with a strip plot to show individual data points
sns.stripplot(x='frailty', y='bmi', data=data, color='grey', alpha=0.6, jitter=True)  # Add jitter for better visibility of individual points

plt.title('BMI by Frailty Status: Summary Statistics and Individual Observations')
plt.xlabel('Frailty Status')
plt.ylabel('BMI')
plt.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey')

# Save the figure to a specified path
plt.savefig('/content/Snaps_of_plots/bmi_vs_frailty_box_strip_plot.png')
plt.show()
```
